```python
import pandas as pd
```

```python

df_a = pd.read_csv("Life Expectancy Data.csv")
df_b = pd.read_csv("life-expectancy.csv")

df_a.columns = df_a.columns.str.lower().str.replace(" ", "_")
df_b.columns = df_b.columns.str.lower().str.replace(" ", "_")

df_b = df_b.rename(columns={"entity": "country", 
                            "period_life_expectancy_at_birth": "life_expectancy"})

df_a["country"] = df_a["country"].str.strip()
df_b["country"] = df_b["country"].str.strip()

df_a["year"] = df_a["year"].astype(int)
df_b["year"] = df_b["year"].astype(int)

df_merged = pd.merge(df_b, df_a, on=["country", "year"], how="left")

df_merged = df_merged.drop(columns=["life_expectancy_"])

aggregates = [
    "Africa","Asia","Europe","Oceania",
    "North America","South America",
    "European Union","World",
    "Low income","Lower-middle income","Upper-middle income","High income",
    "High-income countries","Low-income countries","High-and-upper-middle-income countries",
    "Upper-middle-income countries","Lower-middle-income countries",
    "Central Europe and the Baltics","East Asia & Pacific","Europe & Central Asia",
    "Latin America & Caribbean","Middle East & North Africa","Sub-Saharan Africa","Land-locked Developing Countries (LLDC)",
    "Least developed countries","Least developed countries","Low-and-middle-income countries","Middle-income countries",
    "More developed regions","No income group available","Small Island Developing States (SIDS)","American Samoa","Americas",
    "Low-and-Lower-middle-income countries","Less developed regions, excluding least developed countries","Reunion",
    "Latin America and the Caribbean",
    "Less developed regions",
    "Less developed regions, excluding China",
    "Northern America",
    "Northern Ireland",
    "England and Wales",
    "USSR",
    "Western Sahara",
    "Anguilla","Aruba","Bermuda","Bonaire Sint Eustatius and Saba",
    "British Virgin Islands","Cayman Islands","Cook Islands","Curacao",
    "Falkland Islands","Faroe Islands","French Guiana","French Polynesia",
    "Gibraltar","Greenland","Guadeloupe","Guam","Guernsey","Isle of Man",
    "Jersey","Martinique","Mayotte","Montserrat","New Caledonia","Niue",
    "Northern Mariana Islands","Puerto Rico","Saint Barthelemy",
    "Saint Helena","Saint Martin (French part)","Saint Pierre and Miquelon",
    "Scotland","Sint Maarten (Dutch part)","Tokelau",
    "Turks and Caicos Islands","United States Virgin Islands",
    "Wallis and Futuna"
]

df_merged = df_merged[~df_merged["country"].isin(aggregates)]

df_merged.to_csv("merged_life_expectancy_dataset.csv", index=False)

print("Merging complete!")
print("Final shape:", df_merged.shape)
```
